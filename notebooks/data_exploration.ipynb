# Golden-AI Pro: Data Exploration

Exploring XAUUSD data for trading insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

## 1. Load Data

```python
# Load sample data
df = pd.read_csv('../data/xauusd_h1.csv', parse_dates=['time'], index_col='time')
df.head()
```

## 2. Basic Statistics

```python
df.describe()
```

## 3. Price Visualization

```python
fig, axes = plt.subplots(3, 1, figsize=(15, 10))

# Price
axes[0].plot(df.index, df['close'], label='Close Price')
axes[0].set_title('XAUUSD Price')
axes[0].legend()

# Volume
axes[1].bar(df.index, df['volume'], alpha=0.7)
axes[1].set_title('Volume')

# Returns
returns = df['close'].pct_change()
axes[2].plot(df.index, returns, alpha=0.7)
axes[2].set_title('Returns')
axes[2].axhline(y=0, color='r', linestyle='--')

plt.tight_layout()
plt.show()
```

## 4. Distribution Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Returns distribution
axes[0].hist(returns.dropna(), bins=100, alpha=0.7, density=True)
axes[0].set_title('Returns Distribution')
axes[0].set_xlabel('Return')
axes[0].set_ylabel('Density')

# Box plot by hour
df['hour'] = df.index.hour
sns.boxplot(data=df, x='hour', y='close', ax=axes[1])
axes[1].set_title('Price Distribution by Hour')

plt.tight_layout()
plt.show()
```

## 5. Correlation Analysis

```python
# Calculate features
df['returns'] = df['close'].pct_change()
df['volatility'] = df['returns'].rolling(20).std()
df['sma_20'] = df['close'].rolling(20).mean()
df['sma_50'] = df['close'].rolling(50).mean()

# Correlation matrix
corr_cols = ['open', 'high', 'low', 'close', 'volume', 'returns', 'volatility']
corr = df[corr_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```
